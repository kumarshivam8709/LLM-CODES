```python
rumor=0
nonrumor=0
```

```python
prompt = """
            you are an excellent data annotator. you know every incident and have common sense like humans. your aim is
            to annotate the given text data to 2 labels namely
            1) rumor
            2) non-rumor

             """
```

```python
df.head()
```

```python
df.tail()
```

```python
messages = [{"role": "system", "content": prompt}]

for i in range(1838, 1972, 1):
    newmessage = '''
Tweet: ''' + df['tweet'][i] + '''
Sentiment: '''+df['sentiment'][i]+'''
Emotion: '''+df['max_emotion'][i]+'''
Label: '''

    message = newmessage
    if message:
        messages.append({"role": "user", "content": message})
        chat = openai.ChatCompletion.create(
            model="gpt-3.5-turbo", messages=messages
        )
    reply = chat.choices[0].message.content
    messages.pop()
    if reply == 'rumor':
        rumor += 1
    else:
        nonrumor += 1

    print(f"{i}: {reply}")

    # Check if 100 tweets have been processed and introduce a 5-minute sleep
    if i % 100 == 0:
        print("Sleeping for 5 minutes...")
        time.sleep(300)  # 300 seconds = 5 minutes

print("rumor: ", rumor, "nonrumor: ", nonrumor)
```
